### IFIDiagrams

This Jupyter Notebook script applies all operations required to plot the diagrams.

```python
import os
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

#### Definitions

```python
# Get paths of folders organized by cookie types
current_path = os.getcwd()
wafer = pd.read_csv(os.path.join(current_path, "cw_ssim_results2_chocolate_wafer.csv"))
vanilla = pd.read_csv(os.path.join(current_path, "cw_ssim_results2_vanilla_crescent.csv"))
roll = pd.read_csv(os.path.join(current_path, 
"cw_ssim_results2_wafer_roll.csv"))
jelly = pd.read_csv(os.path.join(current_path, 
"cw_ssim_results2_jelly_cookie.csv"))
```

#### Compute metrics and plot

```python
# Define functions for the calculation of differences of values determined by the metrics. Start with CW-SSIM.
def difference_cw_ssim(df):
    results = []
    for row in df.iterrows():
        for row_match in df.iterrows():
            if row[1]['cookie'] == row_match[1]['cookie'] and row[1]['dirt'] == row_match[1]['dirt'] and row[1]['paper'] == row_match[1]['paper'] and row[1]['dirt_size'] == 7:
                results.append(row[1]['cw_ssim'] - row_match[1]['cw_ssim'])
    return results

# Now add Chi Squared.
def difference_chi2(df):
    results = []
    for row in df.iterrows():
        for row_match in df.iterrows():
            if row[1]['cookie'] == row_match[1]['cookie'] and row[1]['dirt'] == row_match[1]['dirt'] and row[1]['paper'] == row_match[1]['paper'] and row[1]['dirt_size'] == 7:
                results.append(row[1]['chi2'] - row_match[1]['chi2'])
    return results

# And finally, add FSIM, too.
def difference_fsim(df):
    results = []
    for row in df.iterrows():
        for row_match in df.iterrows():
            if row[1]['cookie'] == row_match[1]['cookie'] and row[1]['dirt'] == row_match[1]['dirt'] and row[1]['paper'] == row_match[1]['paper'] and row[1]['dirt_size'] == 7:
                results.append(row[1]['fsim'] - row_match[1]['fsim'])
    return results
```

```python
# Call functions for the calculation of differences of values determined by the metrics and plot diagrams.
results_wafer_cw_ssim = difference_cw_ssim(wafer)
results_vanilla_cw_ssim = difference_cw_ssim(vanilla)
results_roll_cw_ssim = difference_cw_ssim(roll)
results_jelly_cw_ssim = difference_cw_ssim(jelly)

# Average the results
median_wafer_cw_ssim = np.mean(results_wafer_cw_ssim)
median_vanilla_cw_ssim = np.mean(results_vanilla_cw_ssim)
median_roll_cw_ssim = np.mean(results_roll_cw_ssim)
median_jelly_cw_ssim = np.mean(results_jelly_cw_ssim)

# Plot all results for comparison
fig, ax = plt.subplots(2, 2, figsize=(10, 10))
ax[0,0].hist(results_wafer_cw_ssim, bins=50, color='blue', alpha=0.7)
ax[0,0].set_title('Chocolate Wafer')
ax[0,0].set_xlabel('Difference in CW-SSIM')
ax[0,0].set_ylabel('Frequency')
ax[0,1].hist(results_vanilla_cw_ssim, bins=50, color='orange', alpha=0.7)
ax[0,1].set_title('Vanilla Crescent')
ax[0,1].set_xlabel('Difference in CW-SSIM')
ax[0,1].set_ylabel('Frequency')
ax[1,0].hist(results_roll_cw_ssim, bins=50, color='green', alpha=0.7)
ax[1,0].set_title('Wafer Roll')
ax[1,0].set_xlabel('Difference in CW-SSIM')
ax[1,0].set_ylabel('Frequency')
ax[1,1].hist(results_jelly_cw_ssim, bins=50, color='red', alpha=0.7)
ax[1,1].set_title('Jelly Cookie')
ax[1,1].set_xlabel('Difference in CW-SSIM')
ax[1,1].set_ylabel('Frequency')
plt.tight_layout()
plt.show()

# Plot mean values for comparison, too
print(f"Mean difference in CW-SSIM for Chocolate Wafer: {median_wafer_cw_ssim}")
print(f"Mean difference in CW-SSIM for Vanilla Crescent: {median_vanilla_cw_ssim}")
print(f"Mean difference in CW-SSIM for Wafer Roll: {median_roll_cw_ssim}")
print(f"Mean difference in CW-SSIM for Jelly Cookie: {median_jelly_cw_ssim}")
```

```python
# Repeat the processing pipeline above using FSIM
results_jelly_fsim = difference_fsim(jelly)
results_wafer_fsim = difference_fsim(wafer)
results_vanilla_fsim = difference_fsim(vanilla)
results_roll_fsim = difference_fsim(roll)

median_jelly_fsim = np.mean(results_jelly_fsim)
median_wafer_fsim = np.mean(results_wafer_fsim)
median_vanilla_fsim = np.mean(results_vanilla_fsim)
median_roll_fsim = np.mean(results_roll_fsim)

fig, ax = plt.subplots(2, 2, figsize=(10, 10))
ax[0,0].hist(results_wafer_fsim, bins=50, color='red', alpha=0.7)
ax[0,0].set_title('Wafer')
ax[0,0].set_xlabel('Difference in FSIM')
ax[0,0].set_ylabel('Frequency')
ax[0,1].hist(results_vanilla_fsim, bins=50, color='blue', alpha=0.7)
ax[0,1].set_title('Vanilla Crescent')   
ax[0,1].set_xlabel('Difference in FSIM')
ax[0,1].set_ylabel('Frequency')
ax[1,0].hist(results_roll_fsim, bins=50, color='green', alpha=0.7)
ax[1,0].set_title('Wafer Roll')
ax[1,0].set_xlabel('Difference in FSIM')
ax[1,0].set_ylabel('Frequency')
ax[1,1].hist(results_jelly_fsim, bins=50, color='orange', alpha=0.7)
ax[1,1].set_title('Jelly Cookie')
ax[1,1].set_xlabel('Difference in FSIM')
ax[1,1].set_ylabel('Frequency')
plt.tight_layout()
plt.show()

print(np.mean(results_jelly_fsim))
print(np.mean(results_wafer_fsim))
print(np.mean(results_vanilla_fsim))
print(np.mean(results_roll_fsim))
```

```python
# And now with Chi Squared...
results_jelly_chi2 = difference_chi2(jelly)
results_wafer_chi2 = difference_chi2(wafer)
results_vanilla_chi2 = difference_chi2(vanilla)
results_roll_chi2 = difference_chi2(roll)

median_jelly_chi2 = np.mean(results_jelly_chi2)
median_wafer_chi2 = np.mean(results_wafer_chi2)
median_vanilla_chi2 = np.mean(results_vanilla_chi2)
median_roll_chi2 = np.mean(results_roll_chi2)

fig, ax = plt.subplots(2, 2, figsize=(10, 10))
ax[0,0].hist(results_wafer_chi2, bins=50, color='red', alpha=0.7)
ax[0,0].set_title('Wafer')
ax[0,0].set_xlabel('Difference in chi2')
ax[0,0].set_ylabel('Frequency')
ax[0,1].hist(results_vanilla_chi2, bins=50, color='blue', alpha=0.7)
ax[0,1].set_title('Vanilla Crescent')   
ax[0,1].set_xlabel('Difference in chi2')
ax[0,1].set_ylabel('Frequency')
ax[1,0].hist(results_roll_chi2, bins=50, color='green', alpha=0.7)
ax[1,0].set_title('Wafer Roll')
ax[1,0].set_xlabel('Difference in chi2')
ax[1,0].set_ylabel('Frequency')
ax[1,1].hist(results_jelly_chi2, bins=50, color='orange', alpha=0.7)
ax[1,1].set_title('Jelly Cookie')
ax[1,1].set_xlabel('Difference in chi2')
ax[1,1].set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Plot histograms of overview
cookies = {"Chocolate Wafer": wafer,
            "Jelly Cookie": jelly,
            "Vanilla Crescent": vanilla,
            "Wafer Roll": roll}

metrics = ["cw_ssim", "fsim", "chi2"]
metrics_overview = ["CW-SSIM", "FSIM", "Chi^2"]
colors = {5: "blue", 7: "red"}
labels = {5: "5mm", 7: "7mm"}

fig, axes = plt.subplots(len(cookies), len(metrics), figsize=(15, 12))
for i, (cookie_name, df) in enumerate(cookies.items()):
    for j, metric in enumerate(metrics):
        for dirt_size in [5, 7]:
            subset = df[df["dirt_size"] == dirt_size][metric]
            mean_val = subset.mean()
            axes[i, j].hist(subset, bins=30, color=colors[dirt_size], alpha=0.7, label=labels[dirt_size])
            axes[i, j].axvline(mean_val, color=colors[dirt_size], linestyle='dashed', linewidth=2, label=f"{labels[dirt_size]} CoG")
        axes[i, j].set_title(f"{cookie_name} - {metrics_overview[j]}")
        axes[i, j].set_xlabel(metric)
        axes[i, j].set_ylabel("Frequency")
        axes[i, j].legend()
plt.tight_layout()
plt.show()
plt.savefig("metrics_overview.svg")
```

```python
# Define function to round and save metrics
def round_and_save_metrics(df, filename, metrics=["cw_ssim", "fsim", "chi2"], decimals=4):
    """
    Rundet die angegebenen Metriken auf die gewünschte Anzahl von Dezimalstellen
    und speichert das Ergebnis als neue CSV-Datei.
    """
    df_rounded = df.copy()
    for metric in metrics:
        if metric in df_rounded.columns:
            df_rounded[metric] = df_rounded[metric].round(decimals)
    df_rounded.to_csv(filename, index=False)
    print(f"Gespeichert: {filename}")

# Example for all cookies
round_and_save_metrics(wafer, "wafer_metrics_rounded.csv")
round_and_save_metrics(vanilla, "vanilla_metrics_rounded.csv")
round_and_save_metrics(roll, "roll_metrics_rounded.csv")
round_and_save_metrics(jelly, "jelly_metrics_rounded.csv")
```

```python
# Plot histograms of every type individually
cookies = {"Chocolate Wafer": wafer,
            "Jelly Cookie": jelly,
            "Vanilla Crescent": vanilla,
            "Wafer Roll": roll}
metrics = ["cw_ssim", "fsim", "chi2"]
metrics_overview = ["CW-SSIM", "FSIM", "Chi^2"]
colors = {5: "blue", 7: "red"}
labels = {5: "5mm", 7: "7mm"}

for idx, metric in enumerate(metrics):
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))
    axes = axes.flatten()
    for i, (cookie_name, df) in enumerate(cookies.items()):
        for dirt_size in [5, 7]:
            subset = df[df["dirt_size"] == dirt_size][metric]
            mean_val = subset.mean()
            axes[i].hist(subset, bins=30, color=colors[dirt_size], alpha=0.7, label=f"{labels[dirt_size]}")
            axes[i].axvline(mean_val, color=colors[dirt_size], linestyle='dashed', linewidth=2, label=f"{labels[dirt_size]} CoG")
        axes[i].set_title(f"{cookie_name} - {metrics_overview[idx]}", fontsize=16)
        axes[i].set_xlabel(metric, fontsize=14)
        axes[i].tick_params(axis= 'x', labelsize=14)
        axes[i].tick_params(axis = 'y', labelsize=14)
        axes[i].set_ylabel("Frequency", fontsize=14)
        axes[i].legend(fontsize=12)
    plt.tight_layout()
    plt.savefig(f"metrics_{metric}.png")
    plt.show()
```
